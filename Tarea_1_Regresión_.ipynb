```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
import csv
f=open("dataset.csv")
data=list(csv.reader(f))
data
```

```python
x = np.array([float(a[0]) for a in data])
y = np.array([float(a[1]) for a in data])
```
